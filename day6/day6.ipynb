```
file = open("input6.txt", "r")
lines = file.readlines()
print(lines)
```

```
def getGroups(list_):
  group = []
  while '\n' in list_:
    pos = list_.index('\n')
    group.append(list_[:pos])
    list_ = list_[pos+1:]
  group.append(list_)
  return group

groups = getGroups(lines)
```

```
def yesForGroup(group):
  ans = set([])
  for x in group:
    for y in x:
      if y != "\n":
        ans.add(y)  
  return len(ans)

import string


def yesAllGroup(group):
  sol = set(list(string.ascii_lowercase))
  for x in group:
    temp = set([])
    for y in x.rstrip():
      temp.add(y)
    sol = sol.intersection(temp)
  return len(sol)
              
```

```
def sumOfYes(groups):
  sol = 0
  for gr in groups:
    sol += yesForGroup(gr)
  return sol
def sumOfAllYes(groups):
  sol = 0
  for gr in groups:
    sol += yesAllGroup(gr)
  return sol
print("star 1: " ,sumOfYes(groups))
print("star 2: " ,sumOfAllYes(groups))
```
